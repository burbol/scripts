```python
# We store all the density profiles data in the dictionaries water_data and surf_data
sams = [0, 5, 11, 17, 33, 50, 66]
water_data = {} # Variable holds water density data for all the surfaces 
                #-> For Example: water_data[33][0] is the z coordinate and water_data[33][1] is the density of each z coordinate (for sam 33%)
surf_data = {} # Variable holds sam density data for all the surfaces
xvg = gromacs.formats.XVG()
for i in sams:
    xvg.read("dens_NPT_sam"+str(i)+"_water_ptensor.xvg")
    water_data[i] = xvg.array
    xvg.read("dens2_NPT_sam"+str(i)+"_water_ptensor.xvg")
    surf_data[i] = xvg.array
```

```python
# we define the function bulk_dens that calculates the bulk density of water
def bulk_dens(start, end, z, dens_data): #start and end are two float numbers that delimit the bulk zone, z is a vector with the 
                                         # z coordinates, and dens_data another vector with the density values
        
    start_num=round(start/z[1]) #we determine first the z coordinates nearest to the given start and end positions of the interval
    end_num=round(end/z[1])
    
    bulk_value=0.0 #this variable will hold the sum of the density values inside the interval
    points=0 #this variable will hold the number of points inside the interval
    
    for k in range(1000): # loop through all the data points
        if k > start_num and k < end_num: #take only data point inside interval
            #j = dens_data[k]
            bulk_value = bulk_value + dens_data[k]
            points = points + 1
    bulk = bulk_value/points
    return bulk
```

```python
# We store all the bulk values of water in the dictionary bulk_water

start = 3.5  #first we define the limits of the interval with average bulk density
end = 5.5

bulk_water = {}

for i in sams:
    bulk_water[i] =  bulk_dens(start, end, water_data[i][0], water_data[i][1])
    
print "bulk_water=",bulk_water 
```

```python
#L = {}
#for i in sams:
#    L[i] = water_data[i][0][-1] + water_data[i][0][0]  # L is the simulation box height, thus the last z coordinate + one more interval
#print L

L ={0:6.77155, 5:6.7869, 11:6.67282, 17:6.61698, 33:6.52151, 50:6.49091, 66:6.43758}
# we print the resulting box heights L

# The height of each simualtion box and of each surface are stored in the dictionaries L and H
H = {0:2.132, 5:2.181, 11:2.131, 17:2.1, 33:2.074, 50:2.076, 66:2.029}  # H is the height of each sam (detrmined with VMD                                                                         
                                                                        # subtracting the highest atom position to the lowest)
# THE DICTIONARY top_surf STORES THE HEIGHT SURFACE, which will be added to the depletion layer to determine the GDS position
top_surf={0:2.236, 5:2.274, 11:2.224, 17:2.194, 33:2.172, 50:2.179, 66:2.158} # highest atom position for each SAM

# WE DEFINE TWO DICTIONARIES box_len and bulk_line to plot a line showing the bulk position
box_len={}
bulk_line={}
for i in sams:
    box_len[i] = np.arange(0,L[i],water_data[0][0][1])
    bulk_line[i] = np.ones(len(box_len[i]))*bulk_water[i]
```

```python
# The function GDS_distance that calculates 2 x distance between the GDS of each phase
def GDS_distance(L, H, bulk, z, dens_data): # L is the simulation box height, H the height of the surface, bulk a float number 
                                         # with the bulk density value of water, z is a vector with the z coordinates,
                                         # and dens_data a vector with the density values of water 
    dens_integral = scipy.integrate.simps(dens_data, z)
    GDS_dist = (L - H - dens_integral/bulk)
    return GDS_dist
```

```python
# We store all the depletion layer values of water in the dictionary depletion
depletion_x2 = {} # 2 x depletion layer thickness
depletion = {} # depletion layer thickness

for i in sams:
    depletion_x2[i] = GDS_distance(L[i], H[i], bulk_water[i], water_data[i][0], water_data[i][1])

for i in sams: # In this loop we subtract the the depletion layer at the oppsite side of the box
    if i==0: # For the SAM with 0% OH- coverage both layers are equal
        depletion[i] = 0.5*depletion_x2[i]
    else:
        depletion[i] = depletion_x2[i] - depletion[0]
print "depletion=", depletion
```

```python
# NOW WE PLOT THE RESULTING DEPLETION LAYER THICKNESS
d = np.zeros(len(sams))  # d will store the depletion thickness for easier plotting (array is needed, not dictionary)

m = 0
for i in sams: # loop to save values in d
    d[m] = depletion[i] 
    m = m + 1
    
# We define an exponential function "func" to fit the depletion layer thickness results     
def func(x, a, c, d):
    return a*np.exp(-c*x)+d

# We fit the points (we need arrays instead of lists for fitting)
sams2=array(sams)
d2=array(d)
popt, pcov = curve_fit(func, sams2, d2, [0.25, 0.25, 0.25])
 
x = np.arange(0,67,1)
axis_font = {'fontname':'Arial', 'size':'20'}  # we set a bigger font size for the axes labels

plt.plot(sams, d, 'ro') # plot of the data points
plt.plot(x, func(x, popt[0], popt[1], popt[2]), color='black') # plot of the fitted curve
plt.xlabel('OH- percentage', **axis_font)
plt.ylabel(r'$d$ [nm]', **axis_font)
plt.suptitle('Depletion Layer Thickness vs. OH- Percentage', fontsize=14, fontweight='bold')
#plt.text(20, 0.3, 'SAM 0%, Water 1000 molec.',style='italic', bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})
plt.grid(True)
plt.axis([0, 67, 0, 0.30])
#pylab.savefig(namebaserad)
plt.savefig('depletion_OH_ptensor.png',bbox_inches='tight')
```
