```python
import math
import matplotlib.pyplot as plt
import numpy as np
# For the lengends
import matplotlib.patches as mpatches
```

```python
# LOOP THROUGH SUBPLOTS WITH 1 LEGEND FOR EACH PLOT
# For 10 plots in 5 rows and 2 columns:

# we define the color and text of the legend
red_patch = mpatches.Patch(color='red', label='The red data')

Nrows = 5
Ncolumns = 2
fig, ax = plt.subplots(Ncolumns, Nrows)

#fig.subplots_adjust(bottom=0.5,top=3.7, left=0.0, right=1.5, hspace = 0.2)
fig.subplots_adjust(bottom=0.5,top=2.5, left=0.0, right=1.5, hspace = 0.3)

x=np.radians(range(1000))
y=100*np.sin(x)

for i in range(Ncolumns*Nrows):
    ax = plt.subplot(Nrows, Ncolumns, i+1)
    line1, = ax.plot(x,y, 'r', label="test"+str(i))
    a=i
    plt.title('System %s'%(a))
    
    # Create a legend for the first line.
    first_legend = plt.legend(handles=[line1], loc=1)
    # Add the legend manually to the current Axes.
    ax = plt.gca().add_artist(first_legend)

plt.show()
```
