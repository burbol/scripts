```python
# Copy of ExtendSam.ipynb for testing the box sizes

#!/usr/bin/python

# THIS SCRIPT MAKES SAMS BIGGER. THE NUMBER OF COPIES AND THE SIZE OF THE SAMS 
# HAS TO BE CHANGED MANUALLY FOR EACH SYSTEM
# IT MAKES USE OF THE BASH SCRIPT ExtendSamWithArgs.sh ==> Change samsfolder in script!!!!

import matplotlib.pyplot as plt
import numpy as np
import os
import pylab as pl

#Tesing:
pc =[21]    
#pc=[21,25] # percentages of the SAMs we want to change now
##pc=[33,50] 

Nx={21:10, 25:10, 33:9, 50:10} #Number of unit vetors in sam
Ny={21:10, 25:10, 33:8, 50:10} #Number of unit vetors in sam

# Number of copies:
xtimes={21:4, 25:4, 33:4, 50:4}
ytimes={21:5, 25:5, 33:5, 50:5}

#Changed for testing
a0 = 0.5000
oldx = a0
oldy = a0*np.cos(np.pi/6)

# Size of surface
xlength={21:Nx[21]*oldx, 25:Nx[25]*oldx, 33:Nx[33]*oldx, 50:Nx[50]*oldx}
ylength={21:Nx[21]*oldy, 25:Nx[25]*oldy , 33:Nx[33]*oldy, 50:Nx[50]*oldy}

zBoxSize = "20.00000"  #!!! The box size in z-direction can be changed here!! 
#N=10 #Number of unit vetors in sam

for m in pc:
    newboxx = float(oldx)*xtimes[m]*Nx[m]
    newboxy = float(oldy)*ytimes[m]*Ny[m]
    newboxz = float(zBoxSize) 
    
    #samsfolder = "/Users/burbol/Desktop/scripts/SAM_CREATION/SAMs/NEW" #path to .gro files
    #samsfolder = "/Users/burbol/MEGAsync/scripts/SAM_CREATION/SAMs/NEW/drop_placement" #new path of .gro file
    scriptfolder= "/Users/burbol/MEGAsync/scripts/SAM_CREATION/ExtendSAMs" # path to script ExtendSamWithArgs.sh
  
    # Using test file:
    #samsfolder = scriptfolder
    #startgrofile = 'start%d.gro'%(m, )
    samsfolder = '/Users/burbol2/Desktop'
    startgrofile = 'delete.gro'
    
    
    os.chdir(samsfolder)
    last_line = open(startgrofile, "r").readlines()[-1]
    Line_len = len(last_line)
    print last_line
    print Line_len
    #SEEK_END = 2
    h = open(startgrofile, "r")
    #h.seek(-Line_len, SEEK_END)
    #oldx = str(h.read(Line_len)).split("   ")[1].strip()
    #print oldx
    #h.seek(-Line_len, SEEK_END)
    #oldy = str(h.read(Line_len)).split("   ")[2].strip()
    #print oldy
    #h.seek(-Line_len, SEEK_END)
    #oldz = str(h.read(Line_len)).split("   ")[3].strip()
    #print oldz
    h.close()
    
    
    # From here everything runs automatically...
    
    startgrofileBIG = 'start%dbigbox.gro'%(m, )
    os.system("/usr/local/gromacs/bin/editconf -f %s -o %s -box %8.5f %8.5f %8.5f" %(startgrofile, startgrofileBIG, newboxx, newboxy,newboxz, ))
    
    os.chdir(scriptfolder)
    #now we call an external bash script that creates the shifted copies of our gro file 
    #and stores them in the folder "foldername", which is created inside "samsfolder"
    foldername="%d%s"%(m,'reshape', )
    os.system("./ExtendSamWithArgs.sh %s %s %s %s %s %s" %(startgrofileBIG, str(m), str(xtimes[m]), str(xlength[m]), str(ytimes[m]), str(ylength[m]), )) 
    
    def replace_line(file_name, line_num, text):
        lines = open(file_name, 'r').readlines()
        lines[line_num] = text
        out = open(file_name, 'w')
        out.writelines(lines)
        out.close()
    
    def linecount(file_name):
        with open(file_name, 'r') as foo:
            x = len(foo.readlines())
            return x
        
    #we have to count the lines in the starting small gro file
    os.chdir(samsfolder)
    os.chdir(foldername)
    startfile = '%d%dstart%dbigbox.gro'%(0,0,m, )
    linenum = linecount(startfile)
    
    
    #os.chdir(samsfolder)
    #foldername = '%dreshapeBIGGER'%(m, )  
    #os.chdir(foldername, ))
    
    # we create the new file by openning and closing it
    # we will copy all the shifted gro files in newfile 
    newfile = '%sstart%d.gro'%('BIG',m, )
    creating = open(newfile, 'w+')
    creating.close()
        
        
    k = 0
    l = 0
    for i in range(0, xtimes[m]):
        for j in range(0, ytimes[m]):
            filename = '%d%dstart%dbigbox.gro'%(i,j,m, )
            with open(newfile, 'a') as f1:
                with open(filename, 'r') as f:
                    #f1.write('i=' +str(i)+ 'j=' +str(j) +'\n')  
                    for line in f:
                        if i==0 and j==0:
                            if l < linenum-1:
                                #f1.write('i=' +str(i)+ 'j=' +str(j) + line)  
                                f1.write(line)
                                l = l+1
                            elif l == linenum-1:
                                k = 0
                                l = 0
                                break
                        else:
                            #print "l=",l,"((linenum-3)*(k-1))+2=",((linenum-3)*(k-1))+2,"(linenum*k)-1=",(linenum*k)-1
                            if l > (linenum*(k-1))+1 and l < (linenum*k)-1:
                                #f1.write('i=' +str(i)+ 'j=' +str(j) + line)  
                                f1.write(line)
                            #print "l is ",l, "i=",i,"j=",j
                            #if (i==1 and j==6) and l==60683:
                            if (i==(xtimes[m]-1) and j==(ytimes[m]-1)) and l==((xtimes[m]*ytimes[m]-1)*linenum)-1:
                                #f1.write('i=' +str(i)+ 'j=' +str(j) + line)  
                                f1.write(line)
                            l = l + 1
            k = k + 1
        
    #replace molecule number
    molecnum = (linenum-3)*xtimes[m]*ytimes[m]
    replace_line(newfile, 1, str(molecnum) + "\n")
    os.chdir(samsfolder)
    second_line = open(startgrofile, "r").readlines()[1]
    molecnum = int(second_line)
    newmolecnum = molecnum*xtimes[m]*ytimes[m]
    
    #os.system("cd /Users/burbol/Desktop/scripts/SAM_CREATION/SAMs/NEW/21reshape")
    #os.system("source /usr/local/gromacs/bin/GMXRC")
    os.chdir(samsfolder +'/'+ foldername)
    os.system("/usr/local/gromacs/bin/editconf -f %s -o %s -c" %(newfile, newfile, ))
    os.chdir(samsfolder)
    os.system("mv %s  %s"% (startgrofileBIG,samsfolder +'/'+ foldername +'/'+startgrofileBIG ))
    os.chdir(samsfolder +'/'+ foldername)
    os.system("mv %s  %s"% (newfile,'../'+newfile ))
```
