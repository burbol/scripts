```python
#!/usr/bin/env python

# !!IMPORTANT!! If files don't need to be copied, last PRINTED line should be a comment (it's a "cp" command to copy the whole simulations folder)



import os

import math

import fnmatch # for filtering the file list to get the top,gro,mdp,ndx files.



Dir = '.' # directory to save script 

SimDir = '/home/eixeres/files_for_laila/s17_w1000'   #directory from where to run simulation

BackupDir = '/home/eixeres/files_for_laila/s17_w1000'   # directory to copy (backup) simulation output

Filename = 'NVT_sam17_water1000_next'  # name of files
Scriptname = 's17_w1000_next'  # name of script (when NOT testing, set equal to "Filename")

Jobname = 's17_w1000_next' # Name of job


N = 1  # NUMBER OF SCRIPTS TO SUBMIT = total time running will be this number times the Walltime


Walltime = '24:00:00'  # Walltime in format 'h:mm:ss'
SimTime = 25   # maximum run length of simulation in hours, it can be also a fraction of an hour

Email = 'laila.e@fu-berlin.de'

NodesNum = 2  # soroban = 12 cpus/node ; sheldon-ng = 8 cpus/node

CpuNum = 12  # ratio cpus/node!!!

Memory = 2048

SimLen = 60000  # total simulation time length in ps


#next 3 lines only useful when runnig also grompp
#IndexFile = ' '

#FirstGro = 'sam5_water216'

#FirstMdp = 'Mini.mdp'




for i in range(N):

    print('#!/bin/bash\n')

    print('#PBS -N ' + Jobname  + str(i) +  '\n')

    print('#PBS -o ' + Scriptname + '.stdout\n')
    
    print('#PBS -e ' + Scriptname + '.stderr\n')

    print('#PBS -M ' + Email + '\n')

    print('\n')

    print('#PBS -l ppn==' + str(CpuNum)+'\n')

    print('#PBS-l nodes=' + str(NodesNum) +'\n')

    print('#PBS -l walltime=' + Walltime  + '\n')

    print('\n')

    print('cd $PBS_O_WORKDIR \n')

    print('NPROCS=`wc -l < $PBS_NODEFILE`\n')

    print('\n')

    

    # first jobscript changes "mdp" options with tpbconv to extend simulation

    if i == 0:

        print('tpbconv -s ' + Filename + '.tpr  -until ' + str(SimLen) + ' -o ' + Filename + '_next.tpr \n')

        print('mpirun -np ' + str(CpuNum) + ' mdrun -s ' + Filename + '_next.tpr -cpi  '+ Filename + '.cpt -maxh ' +

    	    str(SimTime) + '\n')

    else:

        print('mpirun -np ' + str(CpuNum) + ' mdrun -s ' + Filename + '.tpr -cpi  '+ Filename + '.cpt -maxh ' +

    	    str(SimTime) + '\n')

    	    
    print('\n')
    
    if i < N-1:
    
    	print('qsub ' + Scriptname + str(i+1) + '\n')
    
    elif i == N-1:

        print('#cp -r ' + SimDir + ' ' + BackupDir + '\n')
```
