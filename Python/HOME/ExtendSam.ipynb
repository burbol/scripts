```python
#!/usr/bin/python

# THIS SCRIPT MAKES SAMS BIGGER. THE NUMBER OF COPIES AND THE SIZE OF THE SAMS 
# HAS TO BE CHANGED MANUALLY FOR EACH SYSTEM
# IT MAKES USE OF THE BASH SCRIPT ExtendSamWithArgs.sh ==> Change samsfolder in script!!!!

import matplotlib.pyplot as plt
import numpy as np
import os
import pylab as pl

######### FIRST VERSION #######
# pc=21: xtimes = 2, ytimes = 3, xlength = 6.137, ylength = 2.357 
# pc=25: xtimes = 3, ytimes = 6, xlength = 4.137, ylength = 2.357 
# pc=33: xtimes = 3, ytimes = 4, xlength = 4.5, ylength = 3.464 

######### SECOND VERSION #######
# pc=21: xtimes = 2, ytimes = 3, xlength = 5.000, ylength = 4.333 
# pc=25: xtimes = 2, ytimes = 3, xlength = 5.000, ylength = 4.333 

pc = 21
xtimes = 2
ytimes = 3
xlength = 5.000
ylength = 4.333
startgrofile = 'start%d.gro'%(pc, )
zBoxSize = "12.00000"  #!!! The box size in z-direction can be changed here!! 
#samsfolder = "/Users/burbol/Desktop/scripts/SAM_CREATION/SAMs/NEW" #path to .gro files
samsfolder = "/Users/burbol/Desktop/scripts/SAM_CREATION/SAMs/NEW/drop_placement" #new path of .gro file
scriptfolder= "/Users/burbol/Desktop/scripts/SAM_CREATION/ExtendSAMs" # path to script ExtendSamWithArgs.sh


# From here everything runs automatically...

os.chdir(samsfolder)
last_line = open(startgrofile, "r").readlines()[-1]
Line_len = len(last_line)
SEEK_END = 2
h = open(startgrofile, "r")
h.seek(-Line_len, SEEK_END)
oldx = str(h.read(Line_len)).split("   ")[1].strip()
#print oldx
h.seek(-Line_len, SEEK_END)
oldy = str(h.read(Line_len)).split("   ")[2].strip()
#print oldy
h.seek(-Line_len, SEEK_END)
oldz = str(h.read(Line_len)).split("   ")[3].strip()
#print oldz
h.close()

newboxx = float(oldx)*xtimes
newboxy = float(oldy)*ytimes
newboxz = float(zBoxSize) 

startgrofileBIG = 'start%dbigbox.gro'%(pc, )
os.system("/usr/local/gromacs/bin/editconf -f %s -o %s -box %8.5f %8.5f %8.5f" %(startgrofile, startgrofileBIG, newboxx, newboxy,newboxz, ))

os.chdir(scriptfolder)
#now we call an external bash script that creates the shifted copies to our gro file 
#and stores them in the folder "foldername", which is created inside "samsfolder"
os.system("./ExtendSamWithArgs.sh %s %s %s %s %s %s" %(startgrofileBIG, str(pc), str(xtimes), str(xlength), str(ytimes), str(ylength), )) 

os.chdir(samsfolder)
os.system("mkdir %dreshape"%(pc, ))
foldername = '%dreshape'%(pc, )
os.chdir(foldername)
newfile = '%sstart%d.gro'%('BIG',pc, )

def replace_line(file_name, line_num, text):
    lines = open(file_name, 'r').readlines()
    lines[line_num] = text
    out = open(file_name, 'w')
    out.writelines(lines)
    out.close()

def linecount(file_name):
    with open(file_name, 'r') as foo:
        x = len(foo.readlines())
        return x
    
startfile = '%d%dstart%dbigbox.gro'%(0,0,pc, )
linenum = linecount(startfile)

# we create the new file by openning and closing it
creating = open(newfile, 'w+')
creating.close()

k = 0
l = 0
for i in range(0, xtimes):
    for j in range(0, ytimes):
        filename = '%d%dstart%dbigbox.gro'%(i,j,pc, )
        with open(newfile, 'a') as f1:
            with open(filename, 'r') as f:
                #f1.write('i=' +str(i)+ 'j=' +str(j) +'\n')  
                for line in f:
                    if i==0 and j==0:
                        if l < linenum-1:
                            #f1.write('i=' +str(i)+ 'j=' +str(j) + line)  
                            f1.write(line)
                            l = l+1
                        elif l == linenum-1:
                            k = 0
                            l = 0
                            break
                    else:
                        #print "l=",l,"((linenum-3)*(k-1))+2=",((linenum-3)*(k-1))+2,"(linenum*k)-1=",(linenum*k)-1
                        if l > (linenum*(k-1))+1 and l < (linenum*k)-1:
                            #f1.write('i=' +str(i)+ 'j=' +str(j) + line)  
                            f1.write(line)
                        #print "l is ",l, "i=",i,"j=",j
                        #if (i==1 and j==6) and l==60683:
                        if (i==(xtimes-1) and j==(ytimes-1)) and l==((xtimes*ytimes-1)*linenum)-1:
                            #f1.write('i=' +str(i)+ 'j=' +str(j) + line)  
                            f1.write(line)
                        l = l + 1
        k = k + 1
    
#replace molecule number
molecnum = (linenum-3)*xtimes*ytimes
replace_line(newfile, 1, str(molecnum) + "\n")
os.chdir(samsfolder)
second_line = open(startgrofile, "r").readlines()[1]
molecnum = int(second_line)
newmolecnum = molecnum*xtimes*ytimes

#os.system("cd /Users/burbol/Desktop/scripts/SAM_CREATION/SAMs/NEW/21reshape")
#os.system("source /usr/local/gromacs/bin/GMXRC")
os.chdir(samsfolder +'/'+ foldername)
os.system("/usr/local/gromacs/bin/editconf -f %s -o %s" %(newfile, newfile, ))
os.chdir(samsfolder)
os.system("mv %s  %s"% (startgrofileBIG,samsfolder +'/'+ foldername +'/'+startgrofileBIG ))
os.chdir(samsfolder +'/'+ foldername)
os.system("mv %s  %s"% (newfile,'../'+newfile ))
```
